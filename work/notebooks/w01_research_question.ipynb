## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision: Which specific pages should the content team prioritize for an immediate refresh?

Action: Allocating writer and editor resources to update and improve those flagged pages.

Cost of a wrong call: Wasting expensive writer hours rewriting pages that don't need it or won't improve, while truly declining pages are ignored and continue to lose organic visibility.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Let's check how many pages are actually showing a downward trend
declining_pages = df[df['trend_direction'] == 'down'].shape[0]
print(f"Number of declining pages requiring potential action: {declining_pages}")
```
